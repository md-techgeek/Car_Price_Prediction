# cleanning

```python
backup = car.copy()
```

1. year has many non-year value

```python
car['year'].str.isnumeric()
```

```python
car = car[car['year'].str.isnumeric()]
```

2. year object to int

```python
car['year'] = car['year'].astype(int)
```

3. Price has 'Ask For Price'
   Removing it using exclude

```python
car = car[car['Price'] != 'Ask For Price']
```

4. Price object to int

```python
car['Price']
```

 5. Replacing comma with empty string
 6. Price object to int

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

7. kms_driven has kms and comma with integers

```python
# Using split for removing spce
# get for store
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car['kms_driven']
```

8. In kms_driven having Petrol value

```python
car = car[car['kms_driven'].str.isnumeric()]
```

9. kms_driven object to int

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

10. fuel_Type has nan values

```python
# Finding nan values from fuel_type
car[car['fuel_type'].isna()]
```

```python
# excluding nan using tild sign
car = car[~car['fuel_type'].isna()]
```

11. Name is inconsistant thats the reason keep 1st 3 word of name

```python
# seperating word using split split method
# slicing 1st 3 word using slice method
# and joining it using join
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```


# Model

```python
# Extracting features on x axis
# ALL coloumn are features except PRICE column
x = car.drop(columns = 'Price')
y = car['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name','company','fuel_type'])
                                       ,remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
# Measuring a r2 score
r2_score(y_test,y_pred)
```

```python
# Finding maximum r2 score
scores = []
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = i )
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = np.argmax(scores) )
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```
